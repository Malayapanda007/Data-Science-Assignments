```python
import pandas as pd
import numpy as np
import scipy.stats as st
```

```python
df=pd.read_csv('/content/sales_data_with_discounts.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
l = []
for i in df.columns:
    if pd.api.types.is_numeric_dtype(df[i]):
        l.append(i)
```

```python

for i in l:
    print(i, "Mean  :", np.mean(df[i]))
    print(i, "Median :", np.median(df[i]))
    print(i, "Mode  :", st.mode(df[i]))
    print(i, "standard deviation :", np.std(df[i]))


```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
for i in l:
    sns.histplot(df[i])
    plt.title(f"Histogram of {i}")
    plt.show()
```

```python
for i in l:
    sns.boxplot(df[i])
    plt.title(f"Box Plot of {i}")
    plt.show()
```

```python
df.columns
```

```python
l1=[]
for i in df.columns:
  if i  not in l:
    l1.append(i)
```

```python
l1
```

```python
for i in df[l1].drop('Date', axis=1):
    sns.barplot(df[i])
    plt.title(f"Bar Plot of {i}")
    plt.show()
```

```python
def standard(col):
    m = df[col].mean()
    std = df[col].std()
    standardized_col = (df[col] - m) / std
    return standardized_col


```

```python
list=[]
for i in l:
  list.append(standard(i))
```
