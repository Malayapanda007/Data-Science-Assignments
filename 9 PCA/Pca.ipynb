```python
import matplotlib.pyplot as plt
import seaborn as sns

# Loop through each column in the DataFrame
for col in df.columns:
    # Print descriptive statistics
    print(f"Statistics for '{col}':")
    print(df[col].describe(), "\n")
    
    # Set up the figure with subplots for boxplot, KDE, histogram + KDE, and violin plot
    fig, axes = plt.subplots(1, 4, figsize=(16, 4))  # Reduced figure size (1 row, 4 columns)

    # Box plot (1st plot)
    sns.boxplot(y=df[col], ax=axes[0], color='skyblue')
    axes[0].set_title(f'Box Plot of {col}')
    axes[0].set_ylabel(col)

    # KDE plot (2nd plot)
    sns.kdeplot(df[col], fill=True, ax=axes[1], color='salmon', linewidth=2)
    axes[1].set_title(f'KDE Plot of {col}')
    axes[1].set_xlabel(col)

    # Histogram with KDE (3rd plot)
    sns.histplot(df[col], bins=30, kde=True, ax=axes[2], color='teal')
    axes[2].set_title(f'Histogram + KDE of {col}')
    axes[2].set_xlabel(col)
    axes[2].grid(True, linestyle='--', alpha=0.5)

    # Violin plot (4th plot)
    sns.violinplot(y=df[col], ax=axes[3], palette='Set2')
    axes[3].set_title(f'Violin Plot of {col}')
    axes[3].set_ylabel(col)

    # Adjust layout to make sure plots don't overlap
    plt.tight_layout()
    plt.show()

```

### NoW We Aplly PCA

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
df.head()
```

```python
column=df.columns
```

```python
sc=StandardScaler()
df=sc.fit_transform(df)
```

```python
df=pd.DataFrame(df,columns=column)
df.head()
```

```python
pca=PCA(n_components=None)
pca_df=pca.fit_transform(df)
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.show()
```

```python
Pca=PCA(n_components=3)
pca_df=Pca.fit_transform(df)
```

```python
pca_df=pd.DataFrame(pca_df,columns=['Feature 1','Feature 2','Feature 3'])
pca_df.head()
```

## Kmeans With original Data

```python
wcss = []
for i in range(2, 30):
    km = KMeans(n_clusters=i, random_state=42)
    km.fit(df)
    wcss.append(km.inertia_)

plt.plot(range(2, 30), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
```

```python
plt.plot(wcss)
plt.show()
```

```python
from kneed import KneeLocator
kl = KneeLocator(range(2, 30), wcss, curve='convex', direction='decreasing')
print("Optimal K value (elbow point):", kl.elbow)
```

```python
km= KMeans(n_clusters=11, init='k-means++')
```

```python
pred_cluster=km.fit_predict(df)
pred_cluster
```

```python
df['clustring']=pred_cluster
```

```python
df.head()
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(df,pred_cluster)
```

## Kmeans With PCA Data

```python
wcss = []
for i in range(2, 30):
    km = KMeans(n_clusters=i, random_state=42)
    km.fit(pca_df)
    wcss.append(km.inertia_)

plt.plot(range(2, 30), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
```

```python
from kneed import KneeLocator
kl = KneeLocator(range(2, 30), wcss, curve='convex', direction='decreasing')
print("Optimal K value (elbow point):", kl.elbow)
```

```python
km= KMeans(n_clusters=6, init='k-means++',n_init=13)
```

```python
pred_cluster=km.fit_predict(pca_df)
pred_cluster
```

```python
pca_df['clustring']=pred_cluster
```

```python
pca_df.head()
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(pca_df,pred_cluster)
```
