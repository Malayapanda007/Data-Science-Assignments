```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from statistics import mode
import math
from math import sqrt
```

```python
l=[1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]
```

```python
sample_size = len(l)
sample_mean = np.mean(l)
sample_std = np.std(l)
confidence_level = 0.99
alpha=(1-confidence_level)/2


t_critical = stats.t.ppf(q=confidence_level+alpha, df=sample_size-1)
standard_error=sample_std/(np.sqrt(sample_size))
margine_erreoe=standard_error*t_critical
print(f"CI {sample_mean-margine_erreoe,sample_mean+margine_erreoe}")
```

```python
sample_size = len(l)
sample_mean = np.mean(l)
population_std = 0.2
confidence_level = 0.99
alpha=(1-confidence_level)/2

z_critical = stats.norm.ppf(q=confidence_level+alpha)

standard_error=population_std/(np.sqrt(sample_size))

margine_erreoe=standard_error*z_critical
print(f"CI {sample_mean - margine_erreoe , sample_mean + margine_erreoe}")

```
